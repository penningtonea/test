```python
anatomic_sites = {
    'Head': [],
    'Body': [],
    'Tail': [],
    'Uncinate': []
}

site_comparisons = {}

# Make dict of dataframes labeled as limma numerator and denominator
for subdir in plots.iterdir():
    if subdir.is_dir():
        limma_files = list(subdir.glob("limma_*.csv"))
        for filepath in limma_files:
            numerator = re.sub(r'limma_|_vs.*', '', filepath.stem)
            denominator = re.sub(r'.*vs_', '', filepath.stem) 
            df = pd.read_csv(filepath)

            logfc_col = f'logFC {numerator} VS {denominator}'
    
            # Filter through df for correct p-value and logFC column
            comp_df = (df[df['P.Value'] < 0.01]
                .copy()
                .assign(diffexpressed=lambda df: df[logfc_col]
                        .apply(lambda x: numerator if x > 0 else denominator)))
            
            # Add significant genes to the appropriate list in anatomic_sites dict 
            anatomic_sites[numerator] = comp_df.loc[comp_df['diffexpressed'] == numerator, 'Gene']
            anatomic_sites[denominator] = comp_df.loc[comp_df['diffexpressed'] == denominator, 'Gene']
```
